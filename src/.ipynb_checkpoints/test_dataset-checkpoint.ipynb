```python
import os
from mask_dataset import MaskedDataset
from utils import load_objects
from transformers import AutoTokenizer, BertTokenizer
from tqdm import tqdm
```

```python
# Transform original data frame to tokenized version
# tokenizer = AutoTokenizer.from_pretrained("xlm-roberta-large")
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased')
mlama = MaskedDataset("mlama", "mbert")

from concurrent.futures import as_completed, ProcessPoolExecutor
with ProcessPoolExecutor(max_workers=10) as executor:
    for lang in mlama.langs:
        futures = []
        print("Save tokenized objects for language - {}".format(lang))
        objs = load_objects(lang, "mbert", None)
        for rel in objs.keys():
            future = executor.submit(mlama._save_tokenize_objs, lang, rel, tokenizer)
            futures.append(future)
    for future in as_completed(futures):
        future.result()
```

```python
# Test batchify function
def batchify(sents, batch_size):
    l = len(sents)
    for ndx in range(0, l, batch_size):
        yield sents[ndx:min(ndx + batch_size, l)]
```
