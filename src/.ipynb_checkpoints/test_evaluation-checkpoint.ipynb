```python
# Test read mask-prediction result by language and relation
PREDICTION_ROOT = "../result/prediction"

def tokens2id(pred, tokenizer):
    token_ids = []
    for token in pred:
        assert(len(token) == 1)
        token_ids.append(tokenizer.convert_tokens_to_ids(token)[0])
    return token_ids

def read_pred(lang, rel):
    tgt_fn = os.path.join(PREDICTION_ROOT, lang, "{}-{}.csv".format(lang, rel))
    df = pd.read_csv(tgt_fn)
    df['prediction'] = df['prediction'].apply(lambda x: parse_list(x))
    
    tokenizer = AutoTokenizer.from_pretrained("xlm-roberta-large")
    df['pred_ids'] = df['prediction'].apply(lambda x: tokens2id(x, tokenizer))
    df = df.drop(columns=['Unnamed: 0.1', 'Unnamed: 0'])    
    df.to_csv(tgt_fn)
    return df

df = read_pred(lang, rel)
df.head()
```

```python
# Merge files to folders by languages
import shutil
predict_path = "./prediction"

for lang in mlama.langs:
    shutil.rmtree(os.path.join(predict_path, lang))
    if not os.path.exists(os.path.join(predict_path, lang)):
        os.mkdir(os.path.join(predict_path, lang))

for fn in os.listdir(predict_path):
    target_fn = os.path.join(predict_path, fn)
    if os.path.isfile(target_fn):
        lang = fn.split('-')[0]
        shutil.move(target_fn, os.path.join(predict_path, lang))
```
