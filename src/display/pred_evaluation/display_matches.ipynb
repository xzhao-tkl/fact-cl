```python
import sys
sys.path.append("../../../src")

from mask_dataset import MaskedDataset
dataset = MaskedDataset(model_name='mbert')
```

```python
import random
from pred_evaluation import get_all_and_matched_uuid_lsts
def display_matches(dataset, tgt_lang, display_content=False):
    lang2rel2matched_uuid = get_all_and_matched_uuid_lsts(dataset=dataset, reload=False)[1]
    lang2matched_uuid = {}
    for lang in lang2rel2matched_uuid:
        lang2matched_uuid[lang] = []
        for rel in lang2rel2matched_uuid[lang]:
            if rel == 'P178':
                lang2matched_uuid[lang].extend(lang2rel2matched_uuid[lang][rel])    
    uuids = lang2matched_uuid[tgt_lang]

    print(f"Matched uuids: {len(uuids)}")
    for rel in lang2rel2matched_uuid[lang]:
        if len(lang2rel2matched_uuid[lang][rel]) > 0:
            print(f"\nMatched uuids in relation {dataset.display_rel(rel)} is {len(lang2rel2matched_uuid[lang][rel])}")
            for uuid in random.choices(lang2rel2matched_uuid[lang][rel], k=3):
                print(dataset.display_uuid(uuid))

display_matches(dataset, "en")
```
