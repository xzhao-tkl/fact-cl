```python
import sys
sys.path.append("../../../src")

from mask_dataset import MaskedDataset
dataset = MaskedDataset(model_name='mbert')
```

```python
from pred_evaluation import calculate_overall_p1_score_standard
lang2p1 = calculate_overall_p1_score_standard(dataset)
```

```python
import matplotlib.pyplot as plt

# Sort the dictionary by value in descending order
sorted_data = dict(sorted(lang2p1.items(), key=lambda item: item[1], reverse=True))
lang2p1
# Extracting keys (languages) and values (floats) from the sorted dictionary
languages = [dataset.display_lang(lang, prefix=False) for lang in list(sorted_data.keys())]
values = list(sorted_data.values())

# Plotting the data
plt.figure(figsize=(12, 6))  # Adjust the figure size as needed
plt.bar(languages, values, color='steelblue')
plt.xlabel('Languages', fontsize=18)
plt.ylabel('Precision@1 score', fontsize=18)
plt.title('Factual knowledge probing accuracy',fontsize=20)
plt.xticks(rotation=70, ha='right', fontsize=14)  # Rotate x-labels for better visualization
plt.tight_layout()
plt.show()
```
